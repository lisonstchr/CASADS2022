```python
import pandas as pd
```

```python
# Read the csv file containing the dataframe
df = pd.read_csv('M3_dataset.csv') 
```

```python
# Only select the needed columns

cols = ['PersonNumber', 'IdSession','IdVote', 'GenderAsString', 'Age', 'Region_x', 'Region_y', 'Bildung/Wissenschaft', 'Energie/Umwelt/Land',
       'Finanz/Wirtschaft', 'Medien/Kultur', 'Politik', 'Recht', 'Sozial', 'BGer', 'BK', 'EDA', 'EDI', 'EFD', 'EJPD', 'Parl', 'UVEK', 'VBS', 
        'WBF','Decision' ]
df2= df.loc[:, cols]
```

```python
# Transform the column 'GenderAsString' into a quantitative variable and create a new column 'Male'
def categorise(row):  
    if row['GenderAsString'] == 'm' :
        return 1
    return 0

df2['Male'] = df2.apply(lambda row: categorise(row), axis=1)
df2
```

```python
# Transform the column 'Region_x' into a quantitative variable and split it into 2 columns 'CH-de' and 'CH-fr'
def categorise_regionDE(row):  
    if row['Region_x'] == 'CH-de' :
        return 1
    return 0

df2['CH-de'] = df2.apply(lambda row: categorise_regionDE(row), axis=1)
df2

def categorise_regionFR(row):  
    if row['Region_x'] == 'CH-fr' :
        return 1
    return 0

df2['CH-fr'] = df2.apply(lambda row: categorise_regionFR(row), axis=1)
df2
```

```python
# Transform the column 'Region_y' into a quantitative variable and split it into 2 columns 'Right' and 'Center'
def categorise_PartyRight(row):  
    if row['Region_y'] == 'Right' :
        return 1
    return 0

df2['PartyRight'] = df2.apply(lambda row: categorise_PartyRight(row), axis=1)
df2

def categorise_PartyCenter(row):  
    if row['Region_y'] == 'Center' :
        return 1
    return 0

df2['PartyCenter'] = df2.apply(lambda row: categorise_PartyCenter(row), axis=1)
df2
```

```python
# only select the needed columns

cols = ['PersonNumber', 'IdSession','IdVote', 'Age', 'Bildung/Wissenschaft', 'Energie/Umwelt/Land',
       'Finanz/Wirtschaft', 'Medien/Kultur', 'Politik', 'Recht', 'Sozial', 'BGer', 'BK', 'EDA', 'EDI', 'EFD', 'EJPD', 'Parl', 'UVEK', 'VBS', 
        'WBF', 'Male', 'CH-de', 'CH-fr', 'PartyRight', 'PartyCenter', 'Decision' ]
df3= df2.loc[:, cols]

# Rename the columns

colDict={"PersonNumber":"PersonNumber",
          "IdSession":"IdSession",
          "IdVote":"IdVote",
          "Age":"Age",
          "Bildung/Wissenschaft":"VotingTheme-Bildung/Wissenschaft",
          "Energie/Umwelt/Land":"VotingTheme-Energie/Umwelt/Land",
          "Finanz/Wirtschaft":"VotingTheme-Finanz/Wirtschaft",
          "Medien/Kultur":"VotingTheme-Medien/Kultur",
          "Politik":"VotingTheme-Politik",
          "Recht":"VotingTheme-Recht",
          "Sozial":"VotingTheme-Sozial",
          "BGer":"InitiatingParty-BGer",
          "BK":"InitiatingParty-BK",
          "EDA":"InitiatingParty-EDA",
          "EDI":"InitiatingParty-EDI",
          "EFD":"InitiatingParty-EFD",
          "EJPD":"InitiatingParty-EJPD",
          "Parl":"InitiatingParty-Parl",
          "UVEK":"InitiatingParty-UVEK",
          "VBS":"InitiatingParty-VBS",
          "WBF":"InitiatingParty-WBF",
          "Male":"Gender-Male",
          "CH-de":"Region-CH-de",
          "CH-fr":"Region-CH-fr",
          "PartyRight":"PartyOrientation-Right",
          "PartyCenter":"PartyOrientation-Center",
          "Decision":"Decision(yes=1/no=2/abst=3)",}
df3=df3.rename(columns=colDict)
```

```python
# Change the datatypes of all columns to the coresponding types
# convert all values to integers
df3 = df3.astype(int)
# convert all ID's to strings
df3['PersonNumber'] = df3['PersonNumber'].astype(str)     
df3['IdSession'] = df3['IdSession'].astype(str)   
df3['IdVote'] = df3['IdVote'].astype(str)   
```

```python
#Create the final dataset that contains only integer values and an identifier as index

df3['Identifier'] = df3['PersonNumber'] + '-' + df3['IdSession'] + '-' + df3['IdVote'] #Unique identifier per vote
data = df3.drop(labels=['PersonNumber','IdSession','IdVote'],axis=1) # drop unnecessary labels
data = data.set_index(['Identifier']) # create index
data # final dataframe
```

### 2. Logistic Regression 


```python
# Only select the needed columns for x and y
x = data.drop(labels=['Decision(yes=1/no=2/abst=3)'],axis=1)
y = data['Decision(yes=1/no=2/abst=3)']

# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# Run the logistic regression classification algorithm

for multi_class in ('multinomial', 'ovr'):
    # do fit
    clf = linear_model.LogisticRegression(solver='sag', max_iter=400,
                             multi_class=multi_class, )
    clf.fit(x_train, y_train)

    # print the training scores
    print("training accuracy : %.3f (%s)" % (clf.score(x_train, y_train), multi_class))
    
    # print the test scores
    print("testing accuracy : %.3f (%s)" % (clf.score(x_test, y_test), multi_class))
```

No matter the method (multinomial or ovr), the estimator does not converge, even by increasing the number of iterations up to 400 trials.
It seems that it might not be the best model, as we sould need to increase the number of iteration by a lot to obtain a convergence in estimator. Moreover, as we increased the number of iterations, the accuracy did not increase and stayed at around 65% which is slightly better than pure luck.One way to improve our results could be to get rid of all the abstention votes...

### 3. Trees and Forests

Let's try now a random forest:

```python
# import librairies

from sklearn import ensemble

# Only select the needed columns for x and y
x = data.drop(labels=['Decision(yes=1/no=2/abst=3)'],axis=1)
y = data['Decision(yes=1/no=2/abst=3)']

# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# run the random forest algorithm
for n_est in (1, 4, 50):
    # do fit
    rfc = ensemble.RandomForestClassifier(max_depth=4, n_estimators=n_est,)
    rfc.fit(x_train, y_train)

    # print the training scores
    print("training score : %.3f (n_est=%d)" % (rfc.score(x_train, y_train), n_est))
    # print the test scores
    print("test score : %.3f (n_est=%d)" % (rfc.score(x_test, y_test), n_est))
```

Here again, we only obtain an acuracy of 65%.
What happens if we use a boosted decision tree instead?

```python
# import librairies

from sklearn import ensemble

# Only select the needed columns for x and y
x = data.drop(labels=['Decision(yes=1/no=2/abst=3)'],axis=1)
y = data['Decision(yes=1/no=2/abst=3)']


# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# run the boosted decision tree algorithm
for n_est in (1, 4, 50):
    # do fit
    dtc = ensemble.GradientBoostingClassifier(max_depth=1, n_estimators=n_est,
                                              learning_rate=0.1, subsample=0.5)
    dtc.fit(x_train, y_train)

    # print the training scores
    print("training score : %.3f (n_est=%d)" % (dtc.score(x_train, y_train), n_est))
    # print the test scores
    print("test score : %.3f (n_est=%d)" % (dtc.score(x_test, y_test), n_est))
```

Here again, we only obtain an acuracy of 65%.

### 5. Implementation of Kmeans

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans


# 1. Load the data 
# Only select the needed columns for x and y
x = data.drop(labels=['Decision(yes=1/no=2/abst=3)'],axis=1)
y = data['Decision(yes=1/no=2/abst=3)']

# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# 2. Fix the number of clusters k and define a KMeans clusterer object. Perform the fitting and compute the Silhouette score. 
# Save the results on a list. 
res= []
for n_clusters in range(15,30) :
    clusterer = KMeans(n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(x)
    score= silhouette_score(x,cluster_labels)
    res.append(score)
    
# 3. Plot the Silhouette scores as a function ok k? What is the number of clusters ?
plt.plot(np.arange(len(res))+15, res,'-o')

```

```python

# We could identify clusters, thus we fix the number of clusters as follow : 
n_clusters=24
clusterer = KMeans(n_clusters, random_state=10)
```

```python
clusterer.fit(x_train)
```

```python
cluster_labels=clusterer.predict(x_train)
```

```python
def most_common(nclusters, supervised_labels, cluster_labels):
    """
    Args:
    - nclusters : the number of clusters
    - supervised_labels : for each garment, the labelling provided by the training data ( e.g. in y_train or y_test)
    - cluster_labels : for each garment, the cluster it was assigned by K-Means using the predict method of the Kmeans object
    
    Returns:
    - a list "assignment" of lengths nclusters, where assignment[i] is the majority class of the i-cluster 
    """
    assignment=[]
    for icluster in range(nclusters):
        indices=list(supervised_labels[cluster_labels==icluster])
        try:
            chosen= max(set(indices), key=indices.count)
        except ValueError :
            print('Em')
            chosen=1
        assignment.append(chosen)
    return assignment

assignment=most_common(n_clusters, y_train, cluster_labels)
```

```python
from sklearn.metrics import accuracy_score
y_pred = []
for cluster in clusterer.predict(x_train):
    y_pred.append(assignment[cluster])
y_true = y_train
score = accuracy_score(y_pred, y_true)
```

```python
# print the training scores
print("training score : %.3f" % (score))
```

```python
y_pred = []
for cluster in clusterer.predict(x_test):
    y_pred.append(assignment[cluster])
y_true = y_test
score = accuracy_score(y_pred, y_true)
```

```python
# print the test scores
print("test score : %.3f" % (score))
```

And again, we reach 65% of accuracy....

### 6. Implementation of Multi-layer Perceptron regressor (using sickit learn)

```python
from sklearn.neural_network import MLPClassifier


# 1. Load the data 
# Only select the needed columns for x and y
x = data.drop(labels=['Decision(yes=1/no=2/abst=3)'],axis=1)
y = data['Decision(yes=1/no=2/abst=3)']

# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y\
                                                    , test_size=0.2)


clf = MLPClassifier(hidden_layer_sizes=(100,50), random_state=1\
                    , max_iter=300).fit(x_train, y_train)



# print the training scores
print("training score : %.3f" % (clf.score(x_train, y_train)))
# print the test scores
print("test score : %.3f" % (clf.score(x_test, y_test)))
```
